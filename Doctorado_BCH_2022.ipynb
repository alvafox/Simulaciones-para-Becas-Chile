**Instrucción:** Para Doctorado se quiere igualar la ponderación de AAP a 30% (-5%), para entregar más a la retribución (+5%).

**Hallazgos:** Al redistribuir un 5% del puntaje otorgado a los AAP y destinarlo a la Retribución (2.3), finalmente el número de seleccionados aumenta levemente.

- Es importante remarcar que de 206 seleccionados, se pasa a 211 (+2%).
- Una menor ponderación sobre los AAP favorece el puntaje final de los postulantes, lo que eventualmente podría impactar en puntajes de corte más altos, pero sin tener un efecto "importante" (línea verde en 45º).
- El concurso se vuelve levemente más competitivo, dado que la redistrubución de la ponderación se incorpora a un subcriterio que tiene un desempeño promedio mayor.
- Bajo el supuesto de mantener el puntaje de corte, al realizar la distribución se estarían otorgando más becas.
- Notar que al ser un concurso con un volumen de adjudicación bajo, en comparación a los concursos nacionales, el efecto de las brechas de género en cada comité, pierde relevancia (son números muy pequeños).
- A nivel concurso, la nueva ponderación favorece al género masculino (de 85 a 91 adjudicados), pero de igual forma la adjudicación se mantiene a favor del género femenino (de 121 a 120 adjudicados).

```python
import numpy as np 

df_adm = df_doc[df_doc['ESTADO ADMISIBILIDAD'] != 'FUERA DE BASES']

plt.style.use('ggplot')
# Calcular la correlación entre x e y
corr = df_adm['PUNTAJE_FINAL'].corr(df_adm['PUNTAJE_FINAL_NP'])
m, b = np.polyfit(df_adm['PUNTAJE_FINAL'], df_adm['PUNTAJE_FINAL_NP'], 1)

df_adm.plot(kind='scatter', x='PUNTAJE_FINAL', y='PUNTAJE_FINAL_NP',figsize=(8, 8))
plt.plot(df_adm['PUNTAJE_FINAL'], m*df_adm['PUNTAJE_FINAL'] + b, color='red')
plt.plot([2.5, 5], [2.5,5], color='green') # Lìnea de identidad
plt.title('Correlación entre PUNTAJE_FINAL y PUNTAJE_FINAL_NP (Corr = {:.2f})'.format(corr))
plt.show()
```
